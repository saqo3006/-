```python
import pandas as pd
import matplotlib.pyplot as plt

# Բեռնիր տվյալները
news_df = pd.read_csv("aapl_news.csv")

# Հաշվիր ամենաշատ հոդվածներ ունեցող աղբյուրները
top_sources = news_df['source'].value_counts()


# Կառուցիր bar chart
plt.figure(figsize=(10, 6))
top_sources.plot(kind='bar')
plt.title('Top 10 News Sources by Number of Articles')
plt.xlabel('News Source')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Հաշվիր ամենաշատ հոդվածներ ունեցող աղբյուրները
top_sources1 = news_df['authors'].value_counts().head(20)


# Կառուցիր bar chart
plt.figure(figsize=(10, 6))
top_sources1.plot(kind='bar')
plt.title('Top 20 News Authors by Number of Articles')
plt.xlabel('News Source')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
